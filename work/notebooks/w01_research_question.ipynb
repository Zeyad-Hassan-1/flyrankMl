## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import pandas as pd

# Navigate up to the project root directory so Python can locate the 'data' folder
while not os.path.isdir("data/raw") and os.getcwd() != "/":
    os.chdir("..")

# Load the anonymized starter dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# 1. Calculate the percentage of pages experiencing a downward traffic trend
declining_rate = (df["trend_direction"].str.lower() == "down").mean() * 100

# 2. Calculate the median 90-day impressions for declining pages compared to others
impressions_by_trend = df.groupby("trend_direction")["impressions_90d"].median()

# 3. Check for the Data Gotcha: count pages where avg_position is 0 (which means 'no data', not rank 0)
missing_pos_count = (df["avg_position"] == 0).sum()

print(f"1. Declining pages rate in dataset: {declining_rate:.1f}%")
print(
    f"2. Median 90d impressions for declining pages: {impressions_by_trend.get('down', 0):.0f} (High visibility at stake)"
)
print(
    f"3. Data Gotcha awareness: {missing_pos_count} pages have avg_position = 0 (representing 'no data')"
)
```
